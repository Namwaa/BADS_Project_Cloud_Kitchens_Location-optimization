```python
import webbrowser
import pyautogui
import time
import pyperclip
pyautogui.hotkey('alt','tab')
y = -500
build = []
pyautogui.moveTo(350, 600, duration=0.5)
pyautogui.scroll(-500)
time.sleep(0.25)
for scroll in range(0,64):
    print(scroll)
    x = 340
    for round in range(0,3):
        pyautogui.moveTo(x, 570)
        time.sleep(1)
        pyautogui.click()
        pyautogui.moveTo(300, 60)
        pyautogui.click()
        pyautogui.hotkey('ctrl','c')
        z = pyperclip.paste()
        print(z)
        build.append(z)
        if scroll == 63 and round == 2:
            break
        else:
            pyautogui.hotkey('ctrl','w')
            x +=400
    pyautogui.moveTo(340, 570)
    if scroll%7 != 0 or scroll == 0:
        y = y-9
        pyautogui.scroll(y)
        time.sleep(1)
    else:
        y = 100
        pyautogui.scroll(y)
        y = -500
        pyautogui.scroll(y)
        time.sleep(1)
   
```
